```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Preforming logistic regression and naive bayes on pca dataframe
```

```python
# Loading CSV file
pca_df = pd.read_csv('pca.csv')
```

```python
# Checking information
pca_df.info()
```

```python
# Dropping unnamed column
pca_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
pca_df.head()
```

```python
pca_df.describe().T
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Step 1: Splitting the data into features (X) and target variable (y)
X = pca_df.drop('Label', axis=1)
y = pca_df['Label']

# Step 2: Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.41, random_state=42)

# Step 3: Standardizing features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Fit Logistic Regression model
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train_scaled, y_train)

# Step 5: Evaluate model performance
y_pred = log_reg_model.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
from sklearn.naive_bayes import GaussianNB

# Step 2: Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.41, random_state=42)

# Step 3: Standardizing features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Fit Gaussian Naive Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train)

# Step 5: Evaluate model performance
y_pred_nb = nb_model.predict(X_test_scaled)
print(classification_report(y_test, y_pred_nb))
```

```python
# Calculating accuracy
accuracy_nb = accuracy_score(y_test, y_pred_nb)

print("Naive Bayes Model Accuracy: {:.2f}%".format(accuracy_nb * 100))
```

```python
# While using the pca dataframe that contains the columns with 95% variance:
# Model accuracy for logestic regression came out to be 88.03%.
# Model accuracy for Naive Bayes came out to be 87.77%
```
