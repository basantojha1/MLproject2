# Model Training and Evaluation

```python
# Importing required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading CSV file
pca_df = pd.read_csv('final_pca_data.csv')
```

```python
# Creating a copy of dataframe pca_df
df = pca_df
```

```python
df.head()
```

Preforming one-hot encoding for song and artist columns to use them during modeling

```python
# Performing one-hot encoding on the 'Song' column
Song_encoded = pd.get_dummies(df['Song'], prefix='Song')

# Concatenating the encoded columns with the DataFrame
df = pd.concat([df, Song_encoded], axis=1)

# Drop the original 'Song' column
df.drop('Song', axis=1, inplace=True)
```

```python
# Performing one-hot encoding on the 'Artist' column
Artist_encoded = pd.get_dummies(df['Artist'], prefix='Artist')

# Concatenating the encoded columns with the DataFrame
df = pd.concat([df, Artist_encoded], axis=1)

# Drop the original 'Artist' column
df.drop('Artist', axis=1, inplace=True)
```

```python
# CHecking shape of the dataframe after encoding
df.shape
```

Model Training and Evaluation:

Logistic Regression:

```python
# Importing essential libraries for Logistic Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Data Preparation
X = df.drop('Label', axis=1)
y = df['Label']

# Step 2: Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Step 3: Training the model
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

# Step 4: Evaluating the model
y_pred = logistic_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

Model Accuracy:-
The accuracy is approximately 79.04% meaning the model correctly predicts
the target variable for about 79.04% of instances in the test set.

Model Evaluation:-
For Unpopular:
Precision: 0 (0% of the instances predicted as class 0 were actually class 0)
Recall: 0 (0% of the actual class 0 instances were predicted as class 0)
F1-score: 0 (harmonic mean of precision and recall)
For Popular:
Precision: 0.8 (80% of the instances predicted as class 1 were actually class 1)
Recall: 0.99 (99% of the actual class 1 instances were predicted as class 1)
F1-score: 0.88 (High F1-score indicates better performance in predicting class 1)

```python
# Making predictions on training and testing data
y_train_predict = logistic_model.predict(X_train)
y_test_predict = logistic_model.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, y_train_predict)
test_accuracy = accuracy_score(y_test, y_test_predict)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

Naive Bayes:

```python
# Importing required library for Naive Bayes
from sklearn.naive_bayes import GaussianNB

# Step 1: Data Preparation
X = df.drop('Label', axis=1)
y = df['Label']

# Step 2: Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Step 3: Training the model
naive_bayes_model = GaussianNB()
naive_bayes_model.fit(X_train, y_train)

# Step 4: Evaluating the model
y_pred = naive_bayes_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Making predictions on training and testing data
y_train_predict = naive_bayes_model.predict(X_train)
y_test_predict = naive_bayes_model.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, y_train_predict)
test_accuracy = accuracy_score(y_test, y_test_predict)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

The training accuracy of 1.0 means that your Naive Bayes model has perfectly learned the patterns in the training data. But, testing accuracy of 0.20 is quite low which suggests that the model is not performing well on unseen data. The low testing accuracy indicates that the model is not generalizing well and is not able to make accurate predictions on new data points.

Decision Trees:

```python
# Importing reqired library for Decision Trees
from sklearn.tree import DecisionTreeClassifier

# Step 1: Data Preparation
X = df.drop('Label', axis=1)
y = df['Label']

# Step 2: Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.41, random_state=42)

# Step 3: Training the model
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# Step 4: Evaluating the model
y_pred = decision_tree_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Making predictions on training and testing data
y_train_predict = decision_tree_model.predict(X_train)
y_test_predict = decision_tree_model.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, y_train_predict)
test_accuracy = accuracy_score(y_test, y_test_predict)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

The training accuracy of 1.0 indicates that the decision tree model has perfectly fit the training data, correctly predicting all labels. However, with a testing accuracy of 0.75, the model seems to generalize better to unseen data compared to the Naive Bayes model. This suggests that the decision tree model is more robust and is better for making accurate predictions on new data. 

SVM:

```python
# Importing essential libraries for Support Vector Machine (SVM)
from sklearn.svm import SVC

# Step 1: Data Preparation
X = df.drop('Label', axis=1)
y = df['Label']

# Step 2: Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.31, random_state=42)

# Step 3: Training the model
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# Step 4: Evaluating the model
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Making predictions on training and testing data
y_train_predict = svm_model.predict(X_train)
y_test_predict = svm_model.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, y_train_predict)
test_accuracy = accuracy_score(y_test, y_test_predict)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

The training accuracy of 1.0 indicates that the SVM model has perfectly fit the training data, correctly predicting all labels. However, with a testing accuracy of 0.79, the model seems to generalize better to unseen data compared to the Naive Bayes model and Decision Trees model. This suggests that the decision tree model is more robust and is better for making accurate predictions on new data.
