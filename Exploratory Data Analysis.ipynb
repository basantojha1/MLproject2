# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('billboard_songs_artist.csv')
df2 = pd.read_csv('chorus_features.csv')
```

```python
# Checking for top 5 rows
df1.head()
```

```python
# Checking top 5 rows
df2.head()
```

```python
df2 = df2.drop(columns=['Unnamed: 0'])
```

```python
df2['Song'] = df2['Song'].str.replace('.mp3', '')
```

```python
# Merge df1 and df2 based on the "Song" column
final_df = pd.merge(df1, df2, on='Song', how='left')
```

```python
final_df.head()
```

```python
# Checking for null values in each column
final_df.isnull().sum()
```

```python
# Removing rows with null values
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
# Data Info
final_df.info()
```

```python
# Statistical Summary
final_df.describe()
```

```python
# Specifying the columns for each group
chroma_stft_columns = final_df.loc[:, 'chroma_stft_1':'chroma_stft_84']
chroma_cqt_columns = final_df.loc[:, 'chroma_cqt_1':'chroma_cqt_84']
chroma_cens_columns = final_df.loc[:, 'chroma_cens_1':'chroma_cens_84']
mfcc_columns = final_df.loc[:, 'mfcc_1':'mfcc_140']
rms_columns = final_df.loc[:, 'rms_1':'rms_7']
spectral_centroid_columns = final_df.loc[:, 'spectral_centroid_1':'spectral_centroid_7']
spectral_bandwidth_columns = final_df.loc[:, 'spectral_bandwidth_1':'spectral_bandwidth_7']
spectral_contrast_columns = final_df.loc[:, 'spectral_contrast_1':'spectral_contrast_49']
spectral_rolloff_columns = final_df.loc[:, 'spectral_rolloff_1':'spectral_rolloff_7']
tonnetz_columns = final_df.loc[:, 'tonnetz_1':'tonnetz_42']
zero_crossing_rate_columns = final_df.loc[:, 'zero_crossing_rate_1':'zero_crossing_rate_7']

# Calculating mean for each group
final_df['chroma_stft_mean'] = chroma_stft_columns.mean(axis=1)
final_df['chroma_cqt_mean'] = chroma_cqt_columns.mean(axis=1)
final_df['chroma_cens_mean'] = chroma_cens_columns.mean(axis=1)
final_df['mfcc_mean'] = mfcc_columns.mean(axis=1)
final_df['rms_mean'] = rms_columns.mean(axis=1)
final_df['spectral_centroid_mean'] = spectral_centroid_columns.mean(axis=1)
final_df['spectral_bandwidth_mean'] = spectral_bandwidth_columns.mean(axis=1)
final_df['spectral_contrast_mean'] = spectral_contrast_columns.mean(axis=1)
final_df['spectral_rolloff_mean'] = spectral_rolloff_columns.mean(axis=1)
final_df['tonnetz_mean'] = tonnetz_columns.mean(axis=1)
final_df['zero_crossing_rate_mean'] = zero_crossing_rate_columns.mean(axis=1)
# Now DataFrame contains new columns which contain the means of corresponding columns
```

```python
# Exporting the dataframe to new csv file
final_df.to_csv('final_data.csv')
```

# Bivariate Analysis

```python
# Creating a copy of the dataframe
final_df_copy = final_df.copy()
```

```python
# Renaming labels in the copied DataFrame
final_df_copy['Label'] = final_df_copy['Label'].replace({1: 'Popular', 0: 'Unpopular'})

# Calculating number of popular and unpopular songs
label_counts = final_df_copy['Label'].value_counts()
labels = label_counts.index
sizes = label_counts.values

# Calculating total number of observations
total = sizes.sum()

# Plotting Pie Chart
plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, autopct=lambda p: '{:.0f}'.format(p * total / 100), startangle=140)
plt.title('Number Distribution of Popular and Unpopular Songs')
plt.axis('equal')  # Making equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# Taking artists with highest number of songs in the dataframe
top_artists = final_df['Artist'].value_counts().nlargest(10)

# Plotting bar graph
plt.figure(figsize=(8, 6))
top_artists.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Artists with Highest Number of Songs')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Checking for popular and unpopular song percentage for Taylor Swift
